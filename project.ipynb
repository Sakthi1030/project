```python
import pandas as pd 
```

```python
dates = pd.date_range(start='2021-01-01', end='2021-12-31', periods=500)
dates

```

```python
import random
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
category = [random.choice(categories) for _ in range(500)]


category
```

```python
import numpy as np
likes = np.random.randint(0, 10000, size=500)
```

```python
likes

```

```python
data = dict(zip(['Date', 'Category', 'Likes'], [dates, category, likes]))
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.head(100)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df['Category'].value_counts())
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(subset=['Date', 'Category', 'Likes'] , inplace=True)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
sns.histplot(data=df, x='Likes',color='black')
plt.show()
```

```python
sns.boxplot(data=df, x="Category", y="Likes")
```

```python
print(round(df['Likes'].mean(), 2))
```

```python
group=df.groupby("Category")
mean_likes = group['Likes'].mean()
IX = mean_likes.reset_index()
print(round(IX,2))
```

```python

In this project, we used Python to generate and analyzed random data for a social media experiment by using various methods to
generate random numbers and we ziped the different list to form a data frame
 We used the pandas, numpy, random, and seaborn modules to create, manipulate, and visualize the data.
We also used various methods to clean and summarize the data, such as dropna(), drop_duplicates(), groupby(), mean(), and
value_counts(). 

Some of the key findings from our analysis are:

- The average number of likes for all the posts is 4911.85, with a standard deviation of 2889.12.
- The category with the highest mean of likes is Music, with 5675.62 likes per post, followed by fitness, with 5567.38 
likes per post.
- The category with the lowest mean of likes is Travel, with 4415.21 likes per post .
- The distribution of likes is skewed to the right, with most of the posts having less than 5000 likes, and a few outliers 
having more than 9000 likes.

These findings suggest that the social media users are more interested in topics related to Entertainment and Health such 
as Fashion,music,fitness,andfood, than topics related to tourism and relationships, such as Travel and Family.
They also indicate that there is a high variability in the popularity of the posts, depending on the content and the timing
of the posts. 

This project gives insights into the behavior and preferences of the social media users which can be useful for marketing
strategies
```
