```python
neuron_n=[]
neuron_n.append(int(input("Enter the input layer nodes\t:\t")))
no_of_hidden_layers=int(input("Enter the no of hidden layers\t:\t"))
for i in range(no_of_hidden_layers):
    neuron_n.append(int(input(f"Enter no of neurons in hidden layer {i}")))
neuron_n.append(int(input("Enter no of output layer")))
```

```python
class Node:
    def __init__(self,no,link:dict,output=0):
        self.node_no=no
        self.link=link
        self.output=output
    def update_weights(self,temp_dict,lr:int =1):
        for i,j in self.link.items():
#             print(i,self.link[i],(lr*temp_dict[i]*self.output))
            self.link[i]+=(lr*temp_dict[i]*self.output)
```

```python
print("Enter for node 0 enter 0 links,weights\n eg 0 1,0.6 2,0.5")
node_list={}
for i in neuron_n:
    for j in range(i):
        details=input("Enter node details:").split(" ")
        no=int(details[0])
        dummy_dict={}
        for k in details[1:]:
            print(k)
            link,weight=k.split(",")
            link,weight=int(link),float(weight)
            dummy_dict[link]=weight
        temp_node=Node(no,dummy_dict)
        node_list[no]=temp_node
```

```python
error=100
iterations=0
while abs(error)>0.005:
    iterations+=1
    print(error)
    output={}
    for i,j in node_list.items():
        output[i]=0
    for i,j in input_dict.items():
        node_list[i].output=j
        output[i]+=j
    for i in range(len(node_list)):
        if i>=neuron_n[0]:
            output[i]=sigmoid(output[i])
        for link,weight in node_list[i].link.items():
            output[link]+=weight*output[i]
    print("\t\tOutput\t\t")
    print(output)
    for i in range(len(output)):
        node_list[i].output=output[i]
    error=expected_y-node_list[len(node_list)-1].output
    delta_weights={}
    for i in range(len(node_list)-1,-1,-1):
        if(len(node_list[i].link)==0):
            delta_weights[i]=node_list[i].output*(1-node_list[i].output)*(expected_y-node_list[i].output)
        else:
            dummy=0
            for link,weight in node_list[i].link.items():
                dummy+=(weight*delta_weights[link])
            delta_weights[i]=node_list[i].output*(1-node_list[i].output)*dummy
    print("Delta weights",delta_weights)
    for i,j in node_list.items():
        node_list[i].update_weights(delta_weights,lr)
    print("Updated weights")
    for i,j in node_list.items():
        print(j.node_no,j.link)
```
