```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

```python
sns.histplot(data = tips,x = 'total_bill')
```

```python
sns.scatterplot(x='total_bill',y ='tip',data = tips)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x = 'total_bill',y = 'tip',hue ='sex',size ='size',sizes =(50,100),data = tips)
plt.xlabel('Total_bill')
plt.ylabel('Tips')
plt.title('relationship between total bills and tips')
```

```python
fights = sns.load_dataset('flights')
fights = fights.pivot('month','year','passengers')
sns.heatmap(fights,cmap = 'Blues',annot=True,fmt = 'd')
plt.title('passenger per month')
plt.xlabel('Year')
plt.ylabel('Month')

```
