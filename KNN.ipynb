```python
import pandas as p
result = []
lst = []
lst1=[]
x2=4
y2=6
n = int(input("Enter number of elements : "))

for i in range(0, n):
    ele = int(input())
    lst.append(ele) 

print(lst)
for i in range(0, n):
    ele = int(input())
    lst1.append(ele) 

print(lst1)
for i in range(0,n):
    res=((lst[i]-x2)**2+(lst1[i]-y2)**2)**0.5
    result.append(res)
print(result,3)
data = {'1':lst,
             '2':lst1,
             '3':result}
data=p.DataFrame(data)

rslt_df = data.sort_values(by = '3', ascending = True) 
rslt_df
data['rank'] = range(1, len(data) + 1)
data

```
